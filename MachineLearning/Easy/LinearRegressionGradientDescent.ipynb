```python
import numpy as np
```

```python
"""
Linear regression where model weights are iteratively adjusted to minimize the mean squared erroe between actual and predicted values.
useful for large datasets and when normal equation is computationally expensive

Algorithm
NEGATIVE GRADIENT DIRECTION ; gradient descent updates the weights by moving in the direction that reduces the error most quickly,
using the Loss Function (convert to vector form) 

prediction => y_hat = Xθ
error  => Xθ - y
m => number of samples
Loss => L(θ) = 1/2m (Xθ - y)T (Xθ-y)
gradient => differentiate loss wrt θ, ∇θ​L = 1/m XT(Xθ - y)
gradient descent update rule => θ:=θ−α∇θ​L , α -learning rate ,controls stepsize , too large = overshooting, too small = slow convergenc
"""
```

```python
def LinearRegressionGradientDescent (X,y,alpha,iterations):
    """
    Perform linear regression using gradient descent.

    Args:
        X: Feature matrix of shape (m, n) where first column is all ones (for intercept)
        y: Target vector of shape (m,)
        alpha: Learning rate
        iterations: Number of gradient descent iterations
    
    Returns:
        Learned weights as a 1D array of shape (n,)
    """
    m,n =X.shape
    y = y.reshape(-1,1)
    theta = np.zeros((n,1))

    for _ in range(iterations):
        y_hat = X @ theta # predictions
        e = y_hat - y # error
        g = (1/m)*(np.transpose(X)) @ e
        theta = theta - alpha * g
    return theta.flatten()
```

```python
X = np.array([[1, 1],
              [1, 2],
              [1, 3]])

y = np.array([3, 5, 7])
```

```python
LinearRegressionGradientDescent(X, y, 0.1, 4)
```
