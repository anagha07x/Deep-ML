```python
"""
Linear regression : model the relationship between a scalar dependent variable (y) and one or more explanatory (independent) variable (X).
1. The normal equation provides an analytical solution to find the coefficients (theta) that minimizes the cost function for linear regression.
"""
```

```python
import numpy as np
```

```python
# 1. Using normal equation
# θ = ((XT . X){^-1}).XT.y
def linear_regression_normal_equation (X,y):
    X = np.array(X)
    y = np.array(y).reshape(-1,1)
    x_transpose = np.transpose(X)
    X_variable = (np.linalg.inv(x_transpose @ X))@ x_transpose
    theta =  X_variable @ y
    return np.round(theta,4).flatten().tolist()
```

```python
X = [[1, 1], [1, 2], [1, 3]]
y = [1, 2, 3]
linear_regression_normal_equation (X,y)
```
