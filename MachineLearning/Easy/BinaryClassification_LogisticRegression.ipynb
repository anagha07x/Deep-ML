```python
import numpy as np
```

```python
"""
Given input features and learned parameters (weights and biases) , implement prediction function to compute class probabilities

Takes a batch of samples X (shape: N x D), weights w (shape: D), and bias b
Computes 
z = Xw + b for all samples
Applies the sigmoid function to get probabilities
Returns binary predictions i.e., 0 or 1 using a threshold of 0.5

z => liner combination of features and weights + biases
z = Xw + b
clip z => clip(z,-500,500)
sigmoid = 1/(1+np.exp(-z))
"""
```

```python
def LogisticRegressionBinaryclassification(X,weights,bias):
    X = np.array(X)
    z = X @ weights + bias
    z = np.clip(z,-500,500)
    probabilities = 1/(1+np.exp(-z))
    return (probabilities >= 0.5).astype(int)
```

```python
X = ([[0, 0], [0.1, 0.1], [-0.1, -0.1]])
weights = ([1, 1])
bias = 0
LogisticRegressionBinaryclassification(X,weights,bias)
```
