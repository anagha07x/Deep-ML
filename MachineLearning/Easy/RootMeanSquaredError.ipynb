```python
import numpy as np
```

```python
# ROOT MEAN SQUARED ERROR
def rmse(y_true,y_pred):
    
    y_t = np.array(y_true)
    y_p = np.array(y_pred)

    # empty arrays
    if y_t.size == 0 or y_p.size == 0:
        raise ValueError("arrays can't be empty")

    # mismatched shapes
    if y_t.size != y_p.size:
        raise ValueError("both the arrys must be of same size")

    # valid inputs
    if not np.issubdtype(y_t.dtype,np.number) or not np.issubdtype(y_p.dtype,np.number):
        raise TypeError("the inputs must be number")

    mse = np.mean((y_p - y_t )**2)
    rmse = np.sqrt(mse)

    return round(rmse,3)
```

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
rmse(y_pred,y_true)
```
