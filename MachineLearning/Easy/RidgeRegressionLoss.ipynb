```python
"""
Ridge regression: Linear regression method with a regularization term to revent overfitting by controlling the size of coefficients

1. Regularization: Adds a penalty to losss function to prevent large coefficients to help generalize the model
2. MSE 
3. penalty term: The sum of the squared coefficients, scaled by the regularization parameter λ, which controls the strength of the regularization.

"""
```

```python
import numpy as np
```

```python
def ridge_loss (X,y,w,alpha):
    X = np.array(X)
    w = np.array(w)
    y = np.array(y)
    y_hat = X @ w   # predictions
    mse = np.mean((y - y_hat)**2)
    loss = mse + alpha * np.sum(w**2)
    return loss
```

```python
X = np.array([[1,1],[2,1],[3,1],[4,1]])
w = np.array([.2,2])
y = np.array([2,3,4,5])
alpha = 0.1
ridge_loss (X,y,w,alpha)
```
